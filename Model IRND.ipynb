```python
def Attention_B(X, G, k):
    init = RandomNormal(stddev=0.02)
    theta = Conv2D(k,(2,2), strides = (2,2), padding='same',  kernel_initializer=init)(X)
    Phi = Conv2D(k, (1,1), strides =(1,1), padding='same', use_bias=True)(G)
    ADD = Add()([theta, Phi])
    ADD = Activation('relu')(ADD)
    Psi = Conv2D(1,(1,1), strides = (1,1), padding="same",kernel_initializer=init)(ADD)
    Psi = Activation('sigmoid')(Psi)
    Up = Conv2DTranspose(1, (2,2), strides=(2, 2), padding='valid')(Psi)
    Final = Multiply()([X, Up])
    Final = Conv2D(1, (1,1), strides = (1,1), padding="same",kernel_initializer=init)(Final)
    Final = BatchNormalization(axis=-1, momentum=0.99, epsilon=1e-5)(Final)
    return Final
```

```python
def Unet3(input_shape,n_filters,kernel=(3,3),strides=(1,1),pad='same'):
    conv1 = Conv2D(n_filters,kernel_size=kernel,strides=strides,padding=pad)(input_shape)
    conv1 = BatchNormalization(axis=-1)(conv1)
    conv1 = Activation('relu')(conv1)
    
    conv2 = Conv2D(n_filters,kernel_size=kernel,strides=strides,padding=pad)(conv1)
    conv2 = BatchNormalization(axis=-1)(conv2)
    return Activation('relu')(conv2)
```

```python
def Up3(input1,input2,kernel=(3,3),stride=(1,1), pad='same'):
    up = UpSampling2D(2)(input2)
    up = Concatenate()([up,input1])
    return Unet3(up,int(up.shape[-1]),kernel,stride,pad)
```

```python
def gatingSig(input_shape,n_filters,kernel=(1,1),strides=(1,1),pad='same'):
    conv = Conv2D(n_filters,kernel_size=kernel,strides=strides,padding=pad)(input_shape)
    conv = BatchNormalization(axis=-1)(conv)
    return Activation('relu')(conv)
```

```python
def DSup(x, var):
    d = Conv2D(1,(1,1), strides=(1,1), padding = "same")(x)
    d = UpSampling2D(var)(d)
    return d

def DSup1(x, var):
    d = Conv2D(1,(2,2), strides=(2,2), padding = "same")(x)
    d = UpSampling2D(var)(d)
    return d
```

```python
def RRDB(x,nf=128,gc=32,bias=True):
        x1 = Conv2D(filters=gc, kernel_size=3, strides=1,padding='same', bias=bias)(x)
        x1 = LeakyReLU(alpha=0.25)(x1)
        x2_input = concatenate([x,x1],axis=-1)
        
        x2 = Conv2D(filters= gc, kernel_size=3,strides=1, padding='same',bias=bias)(x2_input)
        x2 = LeakyReLU(alpha=0.25)(x2)
        
        x3_input = concatenate([x,x1,x2] , axis=-1)
        x3 = Conv2D(filters= gc, kernel_size=3,strides=1, padding='same', bias=bias)(x3_input)
        x3 = LeakyReLU(alpha=0.25)(x3)
        
        x4_input = concatenate([x,x1,x2,x3] , axis=-1)
        x4 = Conv2D(filters= gc, kernel_size=3,strides=1, padding='same', bias=bias)(x4_input)
        x4 = LeakyReLU(alpha=0.25)(x4)
        
        x5_input = concatenate([x,x1,x2,x3,x4] , axis=-1)
        x5 = Conv2D(filters= nf, kernel_size=3,strides=1, padding='same', bias=bias)(x5_input)
        x5 = LeakyReLU(alpha=0.25)(x5)
        
        x5 = Lambda(lambda x: x * 0.4)(x5)
        return Add()([x5,x])
    
def RRDB_NET(x,nf=128,gc=32,bias=True):
    out = RRDB(x)
    out = RRDB(out)
    out = RRDB(out)
    out = Lambda(lambda out: out * 0.2)(out)
    return Add()([out,x])

```

```python
def unet(pretrained_weights = None,input_sizea = (128,128,1), input_sizeb = (128,128,1)):

    
    inputimage = Input(input_sizea)
    inputseg = Input(input_sizeb)
    conv1 = Conv2D(16, 3, activation = 'relu', padding = 'same')(inputimage)
    conv1 = Conv2D(16, 3, activation = 'relu', padding = 'same')(conv1)
#     pool1 = MaxPooling2D(pool_size=(2, 2))(conv1)

   
    conv2 = Conv2D(32, 3, activation = 'relu', padding = 'same')(conv1)
    conv2 = Conv2D(32, 3, activation = 'relu', padding = 'same')(conv2)
#     pool2 = MaxPooling2D(pool_size=(2, 2))(conv2)

    conv3 = Conv2D(64, 3, activation = 'relu', padding = 'same')(conv2)
    conv3 = Conv2D(64, 3, activation = 'relu', padding = 'same')(conv3)
#     pool3 = MaxPooling2D(pool_size=(2, 2))(conv3)

    conv4 = Conv2D(1, 3, activation = 'relu', padding = 'same')(inputseg)
    conv41 = Conv2D(1, 3, activation = 'relu', padding = 'same')(conv3)
    
    conv4 = concatenate([conv4, conv41])
    
    pool4 = MaxPooling2D(pool_size=(2, 2))(conv4)
    
    conv4 = Conv2D(128, 3, activation = 'relu', padding = 'same')(conv3)
    pool4 = MaxPooling2D(pool_size=(2, 2))(conv4)
    
    conv4 = Conv2D(256, 3, activation = 'relu', padding = 'same')(pool4)
    pool4 = MaxPooling2D(pool_size=(2, 2))(conv4)
    
#     conv4 = Conv2D(, 3, activation = 'relu', padding = 'same')(pool4)
#     pool4 = MaxPooling2D(pool_size=(2, 2))(conv4)
    
    conv5 = Conv2D(16, 7, activation = 'relu', padding = 'same')(inputimage)
    conv5= Conv2D(16, 7, activation = 'relu', padding = 'same')(conv5)
#     pool1 = MaxPooling2D(pool_size=(2, 2))(conv1)

   
    conv6= Conv2D(32, 7, activation = 'relu', padding = 'same')(conv5)
    conv6= Conv2D(32, 7, activation = 'relu', padding = 'same')(conv6)
#     pool2 = MaxPooling2D(pool_size=(2, 2))(conv2)

    conv7 = Conv2D(64, 7, activation = 'relu', padding = 'same')(conv6)
    conv7 = Conv2D(64, 7, activation = 'relu', padding = 'same')(conv7)
#     pool3 = MaxPooling2D(pool_size=(2, 2))(conv3)
    
    conv8 = Conv2D(1, 7, activation = 'relu', padding = 'same')(inputseg)
    conv81 = Conv2D(1, 7, activation = 'relu', padding = 'same')(conv7)
    
    conv8 = concatenate([conv8, conv81])
    
    pool5 = MaxPooling2D(pool_size=(2, 2))(conv8)
    
    conv8 = Conv2D(128, 7, activation = 'relu', padding = 'same')(pool5)
    pool5 = MaxPooling2D(pool_size=(2, 2))(conv8)
    
    conv8 = Conv2D(256, 7, activation = 'relu', padding = 'same')(pool5)
    pool5 = MaxPooling2D(pool_size=(2, 2))(conv8)
    
#     conv8 = Conv2D(128, 7, activation = 'relu', padding = 'same')(pool5)
#     pool5 = MaxPooling2D(pool_size=(2, 2))(conv8)
    
    
    conv9 = Conv2D(16, 5, activation = 'relu', padding = 'same')(inputimage)
    conv9 = Conv2D(16, 5, activation = 'relu', padding = 'same')(conv9)
#     pool5 = MaxPooling2D(pool_size=(2, 2))(conv5)

   
    conv10 = Conv2D(32, 5, activation = 'relu', padding = 'same')(conv9)
    conv10 = Conv2D(32, 7, activation = 'relu', padding = 'same')(conv10)
#     pool10 = MaxPooling2D(pool_size=(2, 2))(conv10)

    conv10 = Conv2D(64, 5, activation = 'relu', padding = 'same')(conv10)
    conv10 = Conv2D(64, 5, activation = 'relu', padding = 'same')(conv10)
#     pool7 = MaxPooling2D(pool_size=(2, 2))(conv7)
    
    conv11 = Conv2D(1, 5, activation = 'relu', padding = 'same')(inputseg)
    conv111 = Conv2D(1, 5, activation = 'relu', padding = 'same')(conv10)
    
    conv11 = concatenate([conv11, conv111])
    pool6 = MaxPooling2D(pool_size=(2, 2))(conv11)
    
    conv11 = Conv2D(128, 5, activation = 'relu', padding = 'same')(pool6)
    pool6 = MaxPooling2D(pool_size=(2, 2))(conv11)
    
    
    conv11 = Conv2D(256, 5, activation = 'relu', padding = 'same')(pool6)
    pool6 = MaxPooling2D(pool_size=(2, 2))(conv11)
    
#     conv11 = Conv2D(16, 5, activation = 'relu', padding = 'same')(pool6)
#     pool6 = MaxPooling2D(pool_size=(2, 2))(conv11)

    
    DD1 = Flatten()(pool4)
    DD2 = Flatten()(pool5)
    
    DD3 = Flatten()(pool6)
    
    
    drop4 = concatenate([DD1, DD2, DD3])

    DD1 = Activation('relu')(drop4)
    
    
    DD1 = Dense(50, activation='relu', name='dense_post_pool2',  kernel_regularizer=regularizers.l1(0.0001),
                activity_regularizer=regularizers.l2(0.001))(DD1)
    
    DD1 = Dropout(0.5)(DD1)
    
    DD1 = Dense(1, activation='sigmoid', name='predictions',  kernel_regularizer=regularizers.l1(0.0001),
                activity_regularizer=regularizers.l2(0.001))(DD1)
    model = Model(input = [inputimage, inputseg], output = DD1)
    return model
```

```python
def plot_model_history(model_history):
    fig, axs = plt.subplots(1,2,figsize=(15,5))
    axs[0].plot(range(1,len(model_history.history['accuracy'])+1),model_history.history['accuracy'])
    axs[0].plot(range(1,len(model_history.history['val_accuracy'])+1),model_history.history['val_accuracy'])
    axs[0].set_title('Model Accuracy')
    axs[0].set_ylabel('Accuracy')
    axs[0].set_xlabel('Epoch')
    axs[0].set_xticks(np.arange(1,len(model_history.history['accuracy'])+1),len(model_history.history['accuracy'])/10)
    axs[0].legend(['train', 'val'], loc='best')
    # summarize history for loss
    axs[1].plot(range(1,len(model_history.history['loss'])+1),model_history.history['loss'])
    axs[1].plot(range(1,len(model_history.history['val_loss'])+1),model_history.history['val_loss'])
    axs[1].set_title('Model Loss')
    axs[1].set_ylabel('Loss')
    axs[1].set_xlabel('Epoch')
    axs[1].set_xticks(np.arange(1,len(model_history.history['loss'])+1),len(model_history.history['loss'])/10)
    axs[1].legend(['train', 'val'], loc='best')
    plt.show()
```
